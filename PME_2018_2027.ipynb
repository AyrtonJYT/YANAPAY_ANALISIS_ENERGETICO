```python
import pandas as pd
import matplotlib.pyplot as plt

# Re-creación del DataFrame para la independencia del bloque de código
demanda_historica = [2.002, 2.132, 2.223, 2.401, 2.424, 2.642, 2.706, 2.785, 2.768, 2.879, 3.052, 3.207, 3.332, 3.503, 3.670, 3.653, 3.746, 3.933]
demanda_menor_proy = [4.185, 4.785, 5.103, 5.338, 6.399, 6.543, 6.925, 7.228, 7.562]
demanda_medio_proy = [4.295, 4.958, 5.346, 5.661, 6.809, 7.047, 7.533, 7.943, 8.392]
demanda_mayor_proy = [4.385, 5.108, 5.562, 5.954, 7.189, 7.526, 8.123, 8.657, 9.247]

demanda_menor_continua = demanda_historica + demanda_menor_proy
demanda_medio_continua = demanda_historica + demanda_medio_proy
demanda_mayor_continua = demanda_historica + demanda_mayor_proy

data = {
    'Año': list(range(2001, 2028)),
    'Demanda - Menor (MW)': demanda_menor_continua,
    'Demanda - Medio (MW)': demanda_medio_continua,
    'Demanda - Mayor (MW)': demanda_mayor_continua
}

df_demanda_potencia = pd.DataFrame(data).set_index('Año')

# --- Código para generar el gráfico de líneas ---

plt.figure(figsize=(12, 7))

# Graficar las series. En el período 2001-2018, las tres líneas se superponen.
df_demanda_potencia['Demanda - Menor (MW)'].plot(ax=plt.gca(), style='b--', label='Menor (Histórico/Proyección)')
df_demanda_potencia['Demanda - Medio (MW)'].plot(ax=plt.gca(), style='g-', label='Medio (Histórico/Proyección)')
df_demanda_potencia['Demanda - Mayor (MW)'].plot(ax=plt.gca(), style='r-.', label='Mayor (Histórico/Proyección)')

# Marcar el punto de inicio de las proyecciones (separación Histórico/Proyección)
plt.axvline(x=2018.5, color='gray', linestyle=':', linewidth=2, label='Fin Período Histórico (2018)')

# Configuración del gráfico
plt.title('Evolución Continua de la Demanda de Potencia Eléctrica (MW) 2001-2027', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Demanda de Potencia (MW)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Escenario', loc='upper left')
plt.xticks(df_demanda_potencia.index[::2], rotation=45) # Mostrar años de dos en dos para legibilidad
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
